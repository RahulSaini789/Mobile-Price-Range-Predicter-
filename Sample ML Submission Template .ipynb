## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the current dataset and assign it to df
df1 = df.copy()

# Define price range labels
price_ranges = [0, 1, 2, 3]

# Iterate through each price range and count the mobiles
for price_range in price_ranges:
    count = len(df[df['price_range'] == price_range])
    print(f"No. of mobiles belonging to price_range {price_range} is: {count}")
````

From the above description, I can see the minnimum value for columns 'sc_w' (screen width) and for 'px_height' is 0, which is not possible in real life. We need handle these discrepencies in data and will have to replace these values.

````python
# check in how many rows 'sc_w' is 0
df1[df1['sc_w'] == 0].shape[0]
````

````python
# Filter the DataFrame to exclude rows where 'sc_w' is 0
filtered_df = df[df['sc_w'] != 0]

# Create a scatterplot
sns.scatterplot(data=filtered_df, x='sc_h', y='sc_w')
plt.show()
````

As we can see in the above plot, for each value of 'sc_h' there are multiple values of 'sc_w', so our best shot for guessing the values of the 'sc_w' would be taking an aggegate of all available values 'sc_w' for all values of 'sc_h'.

````python
srs = df1[df1['sc_w'] != 0].groupby(by='sc_h')['sc_w'].mean()
srs
````

````python
# ensures that 'sc_w' values are never zero and fills them with corresponding rounded
# 'sc_h' values where necessary.

import math
d = {key: math.ceil(value) for key, value in zip(srs.index, srs.values)}

sc_h = list(df1['sc_h'])
sc_w = list(df1['sc_w'])

for i in range(len(sc_w)):
    if sc_w[i] == 0:
        sc_w[i] = d[sc_h[i]]

df1['sc_h'] = sc_h
df1['sc_w'] = sc_w

# now check the rows with 'sc_w'=0
df1[df1['sc_w'] == 0].shape[0]
````

Converting columns 'sc_h' and 'sc_w' in a single column screen_size , it will be the vertical length of the mobile

````python
df1['screen_size'] = ((df1['sc_h']**2)+(df1['sc_w']**2))**0.5

df1.drop(['sc_h', 'sc_w'], axis=1, inplace=True)
````

````python
# there are only two observations where 'px_height' in 0
df1[df1['px_height'] == 0]
````

Only two rows are showing discrepancy in 'px_heigh', but we'll be using the similar method for this too

````python
filtered_df = df1[df1['px_height'] != 0]

# Create a scatterplot
sns.scatterplot(data=filtered_df, x='px_width', y='px_height')
plt.show()
````

````python
# Let's try linear regression to get this done.
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(df1[df1['px_height'] != 0].loc[:, ['px_width']],
          df1[df1['px_height'] != 0]['px_height'])

model.predict(df1[df1['px_height'] == 0].loc[:, ['px_width']])
````

````python
df1[df1['px_height'] == 0].index
````

````python
# fixing the values
for i, value in zip(df1[df1['px_height'] == 0].index, model.predict(df1[df1['px_height'] == 0].loc[:, ['px_width']])):
    df1.at[i, 'px_height'] = int(value)
````
